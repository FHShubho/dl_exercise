```python
df['mort_acc'].fillna(value=fill_value,inplace=True)
```

```python
df.isnull().sum()
```

```python
#dropping rows with missing values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#droping grade as they are also in sub_grade
df = df.drop('grade',axis=1)
```

```python
df.head()
```

```python
#term into dummy variable
term_dummy = pd.get_dummies(df['term'])
term_dummy
```

```python
#adding dummies to dataframe and removing original column
df = df.drop('term',axis=1)
df = df.join(term_dummy)
df.head()
```

```python
df.columns
```

```python
#sub_grade into dummy variable
grade_dummy = pd.get_dummies(df['sub_grade'])
grade_dummy
```

```python
#adding dummies to dataframe and removing original column
df = df.drop('sub_grade',axis=1)
df = df.join(grade_dummy)
df.head()
```

```python
df.columns
```

```python
#turning into dummies
temp_dummies = pd.get_dummies(df[['verification_status', 'application_type','initial_list_status','purpose' ]],drop_first=True)
temp_dummies
```

```python
#dropping those columns
df = df.drop(['verification_status', 'application_type','initial_list_status','purpose'],axis=1)
df = df.join(temp_dummies)
df.columns
```

```python
df['home_ownership'].value_counts()
```

```python
#merging any and none to others type
df['home_ownership']=df['home_ownership'].replace(['NONE', 'ANY'], 'OTHER')
df['home_ownership'].value_counts()
```

```python
#turning into dummies and removing original column
temp2_dummies = pd.get_dummies(df['home_ownership'],drop_first=True)
df = df.drop('home_ownership',axis=1)
df = df.join(temp2_dummies)
df.columns
```

```python
df['address'].head()
```

```python
#dropping address and issue_d column
df = df.drop('address',axis=1)
df = df.drop('issue_d',axis=1)
```

```python
#getting only the years from cr_line as new column and dropping it
df['earliest_cr_year'] = df['earliest_cr_line'].apply(lambda date:int(date[-4:]))
df = df.drop('earliest_cr_line',axis=1)
df['earliest_cr_year'].head()
```

```python
df.columns
```

```python
df['loan_status'].head()
```

```python
df['loan_status'].unique()
```

```python
#mapping loan status to 0 and 1 
df['loan_status'] = df['loan_status'].map({'Fully Paid': 1, 'Charged Off': 0})
df['loan_status'].head()
```

Training

```python
from sklearn.model_selection import train_test_split
```

```python
#features
X = df.drop('loan_status', axis=1).values
```

```python
#label
y = df['loan_status'].values
```

```python
#25% for test and 75% for training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
#scaling data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#ANN
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Dense(71, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(35, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(17, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=100, batch_size=128, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
#saving model
from tensorflow.keras.models import load_model
model.save('model.h5')
```

Evaluation

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss[['loss','val_loss']].plot()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
prediction = model.predict_classes(X_test)
```

```python
print(classification_report(y_test, prediction))
```

```python
confusion_matrix(y_test, prediction)
```
